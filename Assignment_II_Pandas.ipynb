## Dataset:

```python
PATH_URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/00583/Chemical%20Composion%20of%20Ceramic.csv"

# Save column names as a list
col_ls = ["Ceramic Name", "Part", "Na2O (wt%)","MgO (wt%)","Al2O3 (wt%)", "SiO2 (wt%)", "K2O (wt%)", "CaO (wt%)",
          "TiO2 (wt%)", "Fe2O3 (wt%)", "MnO (mg/L)", "CuO (mg/L)","ZnO (mg/L)","PbO2 (mg/L)","Rb2O (mg/L)","SrO (mg/L)",
          "Y2O3 (mg/L)","ZrO2 (mg/L)","P2O5 (mg/L)"]

df = pd.read_csv(PATH_URL)

# to rename the columns with more informative names
df.columns = col_ls
df.head()
```

#### Attribute Information:

Ceramic.Name: name of ceramic types from Longquan and Jindgezhen

Part: a binary categorical variable ('Body' or 'Glaze')

Na2O: percentage of Na2O (wt%) weight percentage

MgO: percentage of MgO (wt%)

Al2O3: percentage of AI2O3 (wt%)

SiO2: percentage of SiO2 (wt%)

K2O: percentage of K2O (wt%)

CaO: percentage of CaO (wt%)

TiO2: percentage of TiO2 (wt%)

Fe2O3: percentage of Fe2O3 (wt%)

MnO: percentage of MnO (ppm) "parts per million" and it also can be expressed as milligrams per liter (mg/L)

CuO: percentage of CuO (ppm)

ZnO: percentage of ZnO (ppm)

PbO2: percentage of PbO2 (ppm)

Rb2O: percentage of Rb2O (ppm)

SrO: percentage of SrO (ppm)

Y2O3: percentage of Y2O3 (ppm)

ZrO2: percentage of ZrO2 (ppm)

P2O5: percentage of P2O5 (ppm)

#### method 2:

```python
df.groupby(['Part']).mean()
```

A groupby operation involves some combination of splitting the object, applying a function, and combining the results. This can be used to group large amounts of data and compute operations on these groups.

Here I split the dataset into to groups based on the column 'Part', then I applied the function mean, so that it'll calculate the mean for each columns based on each group.

* Research and explain the difference between a pandas Series and a pandas DataFrame.

#### Series
is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.).
#### DataFrame
is a 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dict of Series objects. It is generally the most commonly used pandas object. Like Series, DataFrame accepts many different kinds of input.
